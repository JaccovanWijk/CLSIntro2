```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000
S = [980/n]
I = [20/n]
R = [0]
t = [0]

max_t = 10000

beta = 2
gamma = 0.5

while t[-1] < max_t:
    # Get current values
    current_S = S[-1]
    current_I = I[-1]
    
    # Set rates
    rates = [0, beta * current_S * current_I, gamma * current_I]
    
    rates_sum = sum(rates)
    
    tau = np.random.exponential(scale=1/rates_sum)
    
    t.append(t[-1] + tau)
    
    rand = np.random.random()
    
    if rand * rates_sum < rates[0]:
        S.append(S[-1] + 1/n)
        I.append(I[-1])
        R.append(R[-1])
        
    elif rand * rates_sum > rates[0] and rand * rates_sum < sum(rates[:2]):
        S.append(S[-1] - 1/n)
        I.append(I[-1] + 1/n)
        R.append(R[-1])
        
    elif rand * rates_sum > sum(rates[:2]) and rand * rates_sum < rates_sum:
        S.append(S[-1])
        I.append(I[-1] - 1/n)
        R.append(R[-1] + 1/n)

        
pop = np.add(S, I)
pop = np.add(pop, R)
plt.plot(t,S,label="susceptible")
plt.plot(t,I,label="infected")
plt.plot(t,R,label="recoverd")
plt.plot(t,pop,label="population")
plt.legend()
plt.show()
```

```python
"""Sir with demography"""
n = 1000
S = [980/n]
I = [20/n]
R = [0.0]
t = [0]

max_t = 100000

beta = 2.0
gamma = 0.5
mu =0.1
sigma = 0.25

while t[-1] < max_t:
    # Get current values
    current_S = S[-1]
    current_I = I[-1]
    current_R = R[-1]
    
    # Set rates
    rates = [mu * current_S, beta * current_S * current_I,
             mu * current_I, gamma * current_I, 
             mu * current_R]
    
    rates_sum = sum(rates)
    
    tau = np.random.exponential(scale=1/rates_sum)
    
    t.append(t[-1] + tau)
    
    rand = np.random.random()
    
    # Death
    if rand * rates_sum < rates[0]:
        S.append(S[-1] - 1/n + 1/n)
        I.append(I[-1])
        R.append(R[-1])
    
    # Infection
    elif rand * rates_sum > rates[0] and rand * rates_sum < sum(rates[:2]):
        S.append(S[-1] - 1/n)
        I.append(I[-1] + 1/n)
        R.append(R[-1])
        
    # Death
    elif rand * rates_sum > sum(rates[:2]) and rand * rates_sum < sum(rates[:3]):
        S.append(S[-1] + 1/n)
        I.append(I[-1] - 1/n)
        R.append(R[-1])
    
    # Recovered
    elif rand * rates_sum > sum(rates[:3]) and rand * rates_sum < sum(rates[:4]):
        S.append(S[-1])
        I.append(I[-1] - 1/n)
        R.append(R[-1] + 1/n)
    
    # Death
    elif rand * rates_sum > sum(rates[:4]) and rand * rates_sum < rates_sum:
        S.append(S[-1] + 1/n)
        I.append(I[-1])
        R.append(R[-1] - 1/n)
        
pop = np.add(S, I)
pop = np.add(pop, R)
plt.plot(t,S,label="susceptible")
plt.plot(t,I,label="infected")
plt.plot(t,R,label="recoverd")
plt.plot(t,pop,label="population")
plt.legend()
plt.show()
```

```python
"""Seir"""
n = 1000
S = [980/n]
E = [0]
I = [20/n]
R = [0.0]
t = [0]

max_t = 100000

beta = 2.0
gamma = 0.5
mu =0.1

while t[-1] < max_t:
    # Get current values
    current_S = S[-1]
    current_E = E[-1]
    current_I = I[-1]
    current_R = R[-1]
    
    # Set rates
    rates = [mu * current_S, beta * current_S * current_I, 
             mu * current_E, sigma * current_E,
             mu * current_I, gamma * current_I, 
             mu * current_R]
    
    rates_sum = sum(rates)
    
    tau = np.random.exponential(scale=1/rates_sum)
    
    t.append(t[-1] + tau)
    
    rand = np.random.random()
    
    # Death
    if rand * rates_sum < rates[0]:
        S.append(S[-1] - 1/n + 1/n)
        E.append(E[-1])
        I.append(I[-1])
        R.append(R[-1])
    
    # Exposed
    elif rand * rates_sum > rates[0] and rand * rates_sum < sum(rates[:2]):
        S.append(S[-1] - 1/n)
        E.append(E[-1] + 1/n)
        I.append(I[-1])
        R.append(R[-1])
        
    # Death
    elif rand * rates_sum > sum(rates[:2]) and rand * rates_sum < sum(rates[:3]):
        S.append(S[-1] + 1/n)
        E.append(E[-1] - 1/n)
        I.append(I[-1])
        R.append(R[-1])
    
    # infected
    elif rand * rates_sum > sum(rates[:3]) and rand * rates_sum < sum(rates[:4]):
        S.append(S[-1])
        E.append(E[-1] - 1/n)
        I.append(I[-1] + 1/n)
        R.append(R[-1])
    
    # Death
    elif rand * rates_sum > sum(rates[:4]) and rand * rates_sum < sum(rates[:5]):
        S.append(S[-1] + 1/n)
        E.append(E[-1])
        I.append(I[-1] - 1/n)
        R.append(R[-1])
        
    # Recovered
    elif rand * rates_sum > sum(rates[:5]) and rand * rates_sum < sum(rates[:6]):
        S.append(S[-1])
        E.append(E[-1])
        I.append(I[-1] - 1/n)
        R.append(R[-1] + 1/n)
        
    # Death
    elif rand * rates_sum > sum(rates[:6]) and rand * rates_sum < rates_sum:
        S.append(S[-1] + 1/n)
        E.append(E[-1])
        I.append(I[-1])
        R.append(R[-1] - 1/n)
        
pop = np.add(S, I)
pop = np.add(pop, R)
plt.plot(t,S,label="susceptible")
plt.plot(t,I,label="infected")
plt.plot(t,R,label="recoverd")
plt.plot(t,pop,label="population")
plt.legend()
plt.show()
```
